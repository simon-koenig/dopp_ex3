# Preprocessing of understat.com.csv + Getting first insights

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# import file
with open('./data/understat.com.csv') as f:
    df = pd.read_csv(f)
```

```python
# plot histogramms of all columns
fig,axs = plt.subplots(len(df.columns[4:])//3,3,figsize=(12,20))
i = 4
for triaxis in axs: 
    for ax in triaxis:
        df.hist(column=df.columns[i],ax=ax)
        i+=1
```
